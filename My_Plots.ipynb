```python
import pandas as pd
```

```python
data = pd.read_csv('HotdogSurveyResponses.csv')
data

df=pd.DataFrame(data)


import matplotlib.pyplot as plt
import numpy as np

df
```

```python
filtered_df = df[df['Year at UVA'] != 'Graduate Student']
filtered_df
```

```python
yes_filtered_df = filtered_df[filtered_df['Is a Hotdog a Sandwich?'] == 'Yes']
yes_filtered_df

yes_filtered_df['How strongly do you feel about your previous answer?'].value_counts().sort_index().plot(kind='bar', color='blue', edgecolor='black')

plt.title('Answered "Yes"')
plt.xlabel('Strength of Opinion')
plt.ylabel('Frequency')
plt.show

strength="How strongly do you feel about your previous answer?"
mean_strength=yes_filtered_df[strength].mean()
median_strength=yes_filtered_df[strength].median()
print(f"Mean strength of opinion (Yes)):{mean_strength:.2f}")
print(f"Median strength of opinion (Yes)):{median_strength}")
```

```python
no_filtered_df = filtered_df[filtered_df['Is a Hotdog a Sandwich?'] == 'No']
no_filtered_df

no_filtered_df['How strongly do you feel about your previous answer?'].value_counts().sort_index().plot(kind='bar', color='orange', edgecolor='black')

plt.title('Answered "No"')
plt.xlabel('Strength of Opinion')
plt.ylabel('Frequency')
plt.show

strength="How strongly do you feel about your previous answer?"
mean_strength=no_filtered_df[strength].mean()
median_strength=no_filtered_df[strength].median()
print(f"Mean strength of opinion (No)):{mean_strength:.2f}")
print(f"Median strength of opinion (No)):{median_strength}")
```
